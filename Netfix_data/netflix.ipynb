```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib import style
```

```python
netflix = pd.read_csv("netflix_titles.csv")
```

## Analysis of Movie vs Tv shows

```python
style.use("ggplot")
plt.figure(figsize = (9,5))
plt.hist(netflix["type"] , color = "goldenrod" ,alpha = 0.5,rwidth = 1.5 )
plt.xlabel("Type")
plt.ylabel("Counts")

plt.show()
```

### Observation
    *Data indicates that there are more movies than television shows.

## Movie Rating Analysis

```python
color = [  "yellowgreen", "salmon" , "tomato" , "olive" , "dimgray" , "rosybrown"   ]

plt.bar(netflix["rating"].value_counts().keys() , height = netflix["rating"].value_counts() , color = color )
plt.xticks(rotation = 90)
plt.show()
```

## Top 10 countries of media production

```python
style.use("ggplot")
color = ["dimgray" , "rosybrown" , "salmon" , "tomato" , "olive"]
plt.bar(netflix["country"].value_counts().sort_values(ascending = False)[:10].keys() , height = netflix["country"].value_counts().sort_values(ascending = False)[:10] , color = color)
plt.xticks(rotation = 90)
plt.show()
```

### Observation
    *Above  data indicates that United States create more content and India comes on second number.

# Year wise Analysis

```python
year_wise = netflix["release_year"].value_counts().sort_values(ascending = False)[:10]
```

## Television shows with the most series

```python
netflix_s = netflix[netflix["type"] == "TV Show"]
```

```python
top_10 = netflix_s.sort_values("duration" , ascending = False)
```

```python
plt.figure(figsize = (5,5))
plt.plot(top_10["title"][:10] , top_10["duration"][:10] , color = "blue")
plt.xticks(rotation = 90)
plt.show()
```

## Observation
    *On Netflix, the majority of TV shows have nine seasons, but only one has eight.
